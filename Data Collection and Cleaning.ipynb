```python
# Loading the dataset
import pandas as pd
filepath="C:\Summer Training\Project\Mall_Customers.csv"
data = pd.read_csv(filepath)
# Display the dataset
print(data)

# Checking missing values
missing_values=data.isnull().sum()
print(missing_values)

mean_age=data['Age'].mean()
data["Age"].fillna(mean_age,inplace=True)
print(data)

# Renaming columns for better readability
data.columns = ["CustomerID", "Gender", "Age", "AnnualIncome", "SpendingScore"]
print(data)

# Data transformation (e.g., encoding categorical variables)
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
print(data)
```

```python
'''
1)As there is no missing values, there is no need to clean the data. But I just for a safe side I have replaced the missing values with age column with
their mean.
2)Later I have renamed the columns with some other names  for my convenience. So it would be easy for me while doing work.
3)Then I have encoded the gender variables with 0 for male and 1 for female. it makes our work easy. we can easily identify then.
'''
```
